```python






print("="*60)
print("PYTHON CORE REFRESHER")
print("="*60)




print("\nPART 1 : DATA TYPES")

age = 20
print("\nInteger")
print(age)
print(type(age))

price = 99.99
print("\nFloat")
print(price)
print(type(price))

is_student = True
print("\nBoolean")
print(is_student)
print(type(is_student))

name = "Alice"

print("\nString")
print(name)
print(type(name))

print("Length :", len(name))
print("Upper :", name.upper())
print("Lower :", name.lower())
print("Reverse :", name[::-1])




print("\nLIST")

nums = [10,20,30,40]

print(nums)

print("First element :", nums[0])

nums.append(50)
print("Append :", nums)

nums.pop()

print("After Pop :", nums)

nums.insert(1,100)

print("Insert :", nums)

nums.remove(20)

print("Remove :", nums)

nums.sort()

print("Sorted :", nums)




print("\nTUPLE")

point = (5,7)

print(point)

print(point[0])




print("\nSET")

numbers = {1,2,3,4}

print(numbers)

numbers.add(5)

print(numbers)

numbers.add(5)

print("Duplicate automatically removed")

print(numbers)

print(3 in numbers)




print("\nDICTIONARY")

student = {
    "name":"Alice",
    "age":20,
    "dept":"CSE"
}

print(student)

print(student["name"])

student["age"] = 21

print(student)




print("\n"+"="*60)
print("PART 2 : LOOPS")
print("="*60)

print("\nFOR LOOP")

for i in range(5):
    print(i)

print("\nLoop through list")

nums = [2,4,6,8]

for num in nums:
    print(num)

print("\nUsing enumerate")

for index,value in enumerate(nums):
    print(index,value)

print("\nWHILE LOOP")

i = 0

while i < 5:
    print(i)
    i += 1

print("\nBREAK")

for num in nums:

    if num == 6:
        break

    print(num)

print("\nCONTINUE")

for num in nums:

    if num == 6:
        continue

    print(num)




print("\n"+"="*60)
print("PART 3 : FUNCTIONS")
print("="*60)

def greet():
    print("Welcome to Python!")

greet()


def add(a,b):
    return a+b

answer = add(10,20)

print(answer)


def square(number):
    return number*number

print(square(9))


def even_or_odd(number):

    if number%2==0:
        return "Even"

    return "Odd"

print(even_or_odd(10))
print(even_or_odd(7))




print("\n"+"="*60)
print("PART 4 : LIST COMPREHENSION")
print("="*60)


squares = []

for i in range(6):
    squares.append(i*i)

print(squares)


squares2 = [i*i for i in range(6)]

print(squares2)

evens = [x for x in range(20) if x%2==0]

print(evens)


names = ["Alice","Bob","Charlie"]

lengths = [len(name) for name in names]

print(lengths)


print("\n"+"="*60)
print("PART 5 : DICTIONARY COMPREHENSION")
print("="*60)

square_dict = {x:x*x for x in range(6)}

print(square_dict)


words = ["apple","banana","orange"]

word_lengths = {word:len(word) for word in words}

print(word_lengths)


print("\n"+"="*60)
print("COMMON FUNCTIONS")
print("="*60)

nums = [5,1,9,2,7]

print(nums)

print("Length :",len(nums))

print("Maximum :",max(nums))

print("Minimum :",min(nums))

print("Sum :",sum(nums))

print("Sorted :",sorted(nums))

print("Reverse Sorted :",sorted(nums,reverse=True))


print("\nMEMBERSHIP")

print(5 in nums)

print(100 in nums)


print("\n"+"="*60)
print("MINI DSA EXAMPLES")
print("="*60)


print("\nFind Maximum")

nums = [3,7,1,9,2]

maximum = nums[0]

for num in nums:

    if num > maximum:
        maximum = num

print(maximum)


print("\nFrequency Counter")

nums = [1,2,2,3,3,3,4]

frequency = {}

for num in nums:

    if num in frequency:
        frequency[num] += 1

    else:
        frequency[num] = 1

print(frequency)


print("\nContains Duplicate")

nums = [1,2,3,4,5,2]

seen = set()

duplicate = False

for num in nums:

    if num in seen:
        duplicate = True
        break

    seen.add(num)

print("Contains Duplicate :",duplicate)



print("\nEven Numbers")

nums = [1,2,3,4,5,6,7,8]

evens = []

for num in nums:

    if num%2==0:
        evens.append(num)

print(evens)


print("\nDictionary Example")

students = {
    "Alice":90,
    "Bob":75,
    "Charlie":95
}

for name,marks in students.items():
    print(name,"->",marks)


print("\n"+"="*60)
print("YOU ARE NOW READY TO START LEETCODE!")
print("="*60)
```
